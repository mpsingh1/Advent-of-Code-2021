```python
# file = 'test13.txt'
file = 'input13.txt'
data = []
with open(file) as f:
    lines = f.readlines()
    max_x = 0
    max_y = 0
    for line in lines:
        try:
            int(line[0])
            x, y = line.strip().split(',')
            data.append((int(y),int(x)))
            if int(x)>max_x:
                max_x = int(x)
            if int(y)>max_y:
                max_y = int(y)
        except ValueError:
            continue


print(max_x)
print(max_y)
```

```python
def makePaper(data, size):
    count_dot = 0
    x, y = size
    paper = [[0 for col in range(x)] for row in range(y)]
    for row in range(y):
        for col in range(x):
            if (row, col) in data:
                paper[row][col] = "#"
                count_dot+=1
            else: 
                paper[row][col] = "."
    return paper


def foldY(paper_data, paper, row):
    fold_row = row
#     if row == int(len(paper)/2):
    size_row = (len(paper[0]), fold_row)
    for i, (row, col) in enumerate(data):
        if row > fold_row: 
            paper_data[i] = (row-2*(row-fold_row), col)
    paper_new = makePaper(paper_data, size_row)
    return paper_new, paper_data


def foldX(paper_data, paper, col):
    fold_col = col
    size_col = (fold_col, len(paper))
    for i, (row, col) in enumerate(data):
        if col > fold_col:
            paper_data[i] = (row, col-2*(col-fold_col))
    paper_new = makePaper(paper_data, size_col)
    return paper_new, paper_data


paper_1 = makePaper(data, (max_x+1, max_y+1))
p, d = foldX(data, paper_1, 655)
p, d = foldY(d, p, 447)
p, d = foldX(d, p, 327)
p, d = foldY(d, p, 223)
p, d = foldX(d, p, 163)
p, d = foldY(d, p, 111)
p, d = foldX(d, p, 81)
p, d = foldY(d, p, 55)
p, d = foldX(d, p, 40)
p, d = foldY(d, p, 27)
p, d = foldY(d, p, 13)
p, d = foldY(d, p, 6)
p
```
