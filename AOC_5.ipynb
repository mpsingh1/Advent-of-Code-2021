```python
linspace = max(df.row0.max(), df.row1.max(), df.col0.max(), df.col1.max())+1
matrix = np.zeros((linspace, linspace), dtype='int64')
data_space = pd.DataFrame(matrix)
count = 0

def draw_line(x0 , y0, x1, y1, count):
    global data_space
    if x0 == x1:
        start = min(y0,y1)
        end = max(y0,y1)
        for i in range(start, end+1):
            data_space.iloc[x0,i]+=1
            if data_space.iloc[x0,i] == 2:
                count+=1
                
    elif y0 == y1:
        start = min(x0,x1)
        end = max(x0,x1)
        for i in range(start, end+1):
            data_space.iloc[i,y0]+=1
            if data_space.iloc[i,y0] == 2:
                count+=1
    else: 
        row = int((x1-x0)/abs(x1-x0))
        col = int((y1-y0)/abs(y1-y0))
        for i in range(abs(x1-x0)+1):
            eval_row = x0+row*i
            eval_col = y0+col*i
            data_space.iloc[eval_row, eval_col]+=1
            if data_space.iloc[eval_row, eval_col] == 2:
                count+=1
                
    return count
            

for index, row in df.iterrows():
    count = draw_line(row.row0, row.col0, row.row1, row.col1, count)

print(count)
```
