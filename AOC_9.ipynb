```python
import time
start_time = time.time()
# file = 'test9.txt'
file = 'input9.txt'

with open(file) as f:
    lines = f.readlines()
    total_rows = len(lines)
    total_cols = len(lines[0])-1 #to rid of \n in lines
    data = [[0 for col in range(total_cols)] for row in range(total_rows)]
    
    for row, line in enumerate(lines):
        line = line.strip()
        for col, num in enumerate(line):
            data[row][col] = int(num)

            
def adjacent(point):
    global total_rows
    global total_cols
    row, col = point
    up = (row + 1, col)
    down = (row - 1, col)
    left = (row, col - 1)
    right = (row, col + 1)
    
    all_indices = [up, down, left, right]
    
    valid_indices = [(y, x) for y, x in all_indices if y in range(total_rows) and x in range(total_cols)]
    
    return valid_indices


def isLowPoint(point):
    global data
    row, col = point
    count = 0
    adjacent_points = adjacent(point)
    for point in adjacent_points:
        if data[row][col] >= data[point[0]][point[1]]:
            count+=1
    if count == 0:
        return True


def pointsBasin(points):
    global data
    return_points = []
    for row, col in points:
        point = (row, col)
        adjacent_points = adjacent(point)
        basin_points = [(y, x) for y, x in adjacent_points if data[y][x] > data[row][col] and data[y][x] != 9]
        if len(basin_points) > 0:
            return_points += basin_points
    unique_return_points = list(set(return_points))
    if len(unique_return_points) > 0:
        return True, unique_return_points, len(unique_return_points)
    else: 
        return False, unique_return_points, len(unique_return_points)


# part 1
lowest = []
for row in range(total_rows):
    for col in range(total_cols):
        point = (row, col)
        if isLowPoint(point):
            lowest.append(data[row][col])
            
print(f'Part 1: {sum(lowest)+len(lowest)}')

# part 2
basin_sizes = []

for row in range(total_rows):
    for col in range(total_cols):
        if isLowPoint((row, col)):
            points = [(row, col)]
            basin_points = []
            isBasin = True
            while isBasin:
                isBasin, points, num_points = pointsBasin(points)
                basin_points += points

            basin_sizes.append(len(list(set(basin_points)))+1)

import numpy as np
print(f'Part 2: {np.prod(sorted(basin_sizes)[-3:])}')

print("--- %s seconds ---" % (time.time() - start_time))
```
